```python
# Basic libraries
import pandas as pd
import numpy as np
import pickle
import os

# ML libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load dataset
df = pd.read_csv("../dataset/Diseases_and_Symptoms_dataset.csv")

# Show data
df.head()
```

```python
# Check columns
print(df.columns)

# Fill missing values
df.fillna(0, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# clean column names
df.columns = df.columns.str.strip()

# find target column
target_col = None
for col in df.columns:
    if "disease" in col.lower() or "prognosis" in col.lower():
        target_col = col
        break

print("Detected target column:", target_col)

# split data
X = df.drop(target_col, axis=1)
y = df[target_col]
```

```python
X = df.drop("diseases", axis=1)
y = df["diseases"]
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Create model folder
os.makedirs("../model", exist_ok=True)

# Save model
pickle.dump(model, open("../model/disease_model.pkl", "wb"))

# Save encoder
pickle.dump(le, open("../model/encoder.pkl", "wb"))

# Save column names
pickle.dump(X.columns, open("../model/columns.pkl", "wb"))

print("Model saved successfully!")
```

```python
# Example: random sample
sample = X.iloc[[0]]   # keep it as DataFrame

prediction = model.predict(sample)
disease = le.inverse_transform(prediction)

print("Predicted Disease:", disease[0])
```
