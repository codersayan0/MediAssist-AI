```python
import pandas as pd
import joblib
```

```python
med_df = pd.read_csv("../datasets/medicine/recommendations/medications.csv")
diet_df = pd.read_csv("../datasets/medicine/recommendations/diets.csv")
prec_df = pd.read_csv("../datasets/medicine/recommendations/precautions_df.csv")
workout_df = pd.read_csv("../datasets/medicine/recommendations/workout_df.csv")
```

```python
med_df.columns = med_df.columns.str.lower()
diet_df.columns = diet_df.columns.str.lower()
prec_df.columns = prec_df.columns.str.lower()
workout_df.columns = workout_df.columns.str.lower()
```

```python
recommendation_dict = {}

for disease in med_df["disease"].unique():
    recommendation_dict[disease] = {
        "medicine": med_df[med_df["disease"] == disease]["medication"].values.tolist(),
        "diet": diet_df[diet_df["disease"] == disease]["diet"].values.tolist(),
        "precaution": prec_df[prec_df["disease"] == disease].iloc[:, 1:].values.flatten().tolist(),
        "workout": workout_df[workout_df["disease"] == disease]["workout"].values.tolist()
    }
```

```python
joblib.dump(recommendation_dict, "../models/recommendation.pkl")
```
